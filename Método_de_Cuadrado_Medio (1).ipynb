```python
import pandas as pd

def cuadrado_medio(semilla, n):
    # Variables iniciales
    resultado = []
    k = len(str(semilla))
    r = semilla

    for i in range(n+1):
        rn_cuadrado = r ** 2
        mrn_cuadrado = int(str(rn_cuadrado)[1:-1])
        val1 = int(str(mrn_cuadrado)[:k])
        val2 = int(str(mrn_cuadrado)[-k:])

        resultado.append([ r, rn_cuadrado, mrn_cuadrado, val1, val2])

        if val1 == 0:
            break

        r = val1

    # Crear un DataFrame
    df = pd.DataFrame(resultado, columns=[ "R(n)", "R(n)^2", "MR(n)^2", "Val 1", "Val 2"])

    return df

semilla = 985
n = 30

df = cuadrado_medio(semilla, n)
df
```
