```python
import pandas as pd

df = pd.read_csv('base.csv')


df['SAP_IMRG_DATE_MEAS'] = pd.to_datetime(df['SAP_IMRG_DATE_MEAS'], format='%Y%m%d')
df['Year'] = df['SAP_IMRG_DATE_MEAS'].dt.year
fuel_per_year = df.groupby(['PLANT_PLANT_NAME', 'Year'])['FM_SI_UNIT_CONVERTED'].sum().reset_index()

fuel_per_year['CO2_emissions_kg'] = fuel_per_year['FM_SI_UNIT_CONVERTED'] * 2.7

def calculate_kpi(group):
    group = group.sort_values('Year')
    group['kpi_CO2_reduction'] = group['CO2_emissions_kg'].diff(periods=1)
    return group

result = fuel_per_year.groupby('PLANT_PLANT_NAME').apply(calculate_kpi)
result
```
