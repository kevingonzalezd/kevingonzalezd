```python
import requests

location = '25.651933,-100.289291'  # Distrito Tec

url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={location}&radius=4000&type=transit_station&key={api_key}"

response = requests.get(url)

data = response.json()

Paradas = []

if 'results' in data:
    stops = data['results']
    print(f"Paradas de transporte público cerca del Distrito Tec:")
    for stop in stops:
        lat = stop['geometry']['location']['lat']
        lng = stop['geometry']['location']['lng']

        Paradas.append((lat, lng))

    print("Coordenadas de las paradas:")
    for parada in Paradas:
        print(f"Latitud: {parada[0]}, Longitud: {parada[1]}")
else:
    print("No se encontraron paradas de transporte público cercanas.")
```
