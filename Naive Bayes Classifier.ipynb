```python
## General Libraries
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_csv('political_social_media.csv', encoding='latin-1')
data
```

```python
data_clean = data
data_clean['target'] = data_clean['bias'].map({'neutral' : 0, 'partisan' : 1})
data_clean = data_clean.rename(columns={'text' : 'tweet'})
data_clean=data_clean[['tweet', 'target']]
data_clean.head()
```

```python
data_clean['tweet'] = data_clean['tweet'].apply(lambda x : re.sub('[^A-Za-z0-9 ]+', ' ', x))
data_clean
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
print(stop_words)
```

```python
from nltk.tokenize import word_tokenize
```

```python
def remove_stop_words(message):
    
    words = word_tokenize(message)
    words = [word for word in words if word not in stop_words]
    
    return words
```

```python
import nltk
nltk.download('punkt')

data_clean['tweet'] = data_clean['tweet'].apply(remove_stop_words)
data_clean
```

```python
np.random.seed(1337) 
pruebas=data_clean.iloc[-6:]
data_clean=data_clean.iloc[:-6]

number_of_rows = data_clean.shape[0]
index_train = np.random.choice(range(number_of_rows), int(0.9 * number_of_rows), replace=False)
index_test = np.asarray(list(set(range(number_of_rows)) - set(index_train)))
train_set = data_clean.iloc[index_train] 
test_set = data_clean.iloc[index_test] 
print(train_set.shape)
print(test_set.shape)
```

```python
p_ham = train_set[train_set['target'] == 0].shape[0] / train_set.shape[0]
p_ham
```

```python
def bag_of_words(corpus): # Parametro: DataFrame.column
    
    words = [] # Inicializar lista de palabras
    # Agregar todas las palabras a la lista.
    for element in corpus:
        for word in element:
            words.append(word) 
    
    # Generar datarame con conteo de palabras unicas.
    words = pd.DataFrame(pd.Series(words)).value_counts().rename_axis('words').reset_index(name='count')

    # Convertir dataframe a diccionario
    bag_of_words = dict(zip(words['words'], words['count']))
        
    return bag_of_words, words
```

```python
bag_target, words_target = bag_of_words(train_set[train_set['target'] == 1].tweet)
```

```python
bag_ham, words_ham = bag_of_words(train_set[train_set['target'] == 0].tweet)
```

```python
bag_words, words = bag_of_words(train_set.tweet)
```

```python
def probability_words(df): # Prametro: DataFrame
    
    total = df["count"].sum() # Cantidad total de palabras
        
    probability_words = {} # Inicializar dict: probability_words

    for word in df.iloc[:,0]:
        probability_words[word] = int(df[df.iloc[:,0]==word].iloc[:,1]) / total      
    return probability_words
```

```python
prob_target = probability_words(words_target)
prob_ham = probability_words(words_ham)
prob_words = probability_words(words)
```

```python
def classify_tweet(tweet, p_gh, p_gs, p_w, p_s, p_h):
    prob_ham = prob_target = 1
    for word in tweet:
        if word in p_gs:
            prob_target = prob_target * p_gs[word] / p_w[word]
        if word in p_gh:
            prob_ham = prob_ham * p_gh[word] / p_w[word]
    prob_target = prob_target * p_s
    prob_ham = prob_ham * p_h
    if prob_target > prob_ham:
        return 1
    else: return 0
```

```python
test_set_hat = test_set.copy()
test_set_hat["prediction"] = 1
for i in range(len(test_set_hat.index)):
    test_set_hat.prediction.iloc[i] = int(classify_tweet(test_set_hat.tweet.iloc[i], prob_ham, prob_target, prob_words, p_target, p_ham))
```

```python
def performance_metrics(results):
    
    positives = results[['target', 'prediction']][results['target'] == 1]
    negatives = results[['target', 'prediction']][results['target'] == 0]
    
    true_negatives = negatives[negatives['target'] == negatives['prediction']].shape[0]
    false_positives = negatives[negatives['target'] != negatives['prediction']].shape[0]
    true_positives = positives[positives['target'] == positives['prediction']].shape[0]
    false_negatives = positives[positives['target'] != positives['prediction']].shape[0]
    
    confusion_matrix = {'actual positives' : [true_positives, false_negatives], 
                        'actual negatives' : [false_positives, true_negatives]}
    
    confusion_matrix_df = pd.DataFrame.from_dict(confusion_matrix, orient='index', 
                                                 columns=['predicted positives', 'predicted negatives'])
    
    accuracy = (true_positives + true_negatives) / (true_positives + false_positives +  true_negatives + false_negatives)
    precission = true_positives / (true_positives + false_positives)
    recall = true_positives / (true_positives + false_negatives)
    f1_score = 2 * (precission * recall) / (precission + recall)
    
    metrics = {'Accuracy' : accuracy, 'Precission' : precission, 'Recall' : recall, 'F1 Score' : f1_score}
    
    metrics_df = pd.DataFrame.from_dict(metrics, orient='index', columns=['Metrics'])
    
    return confusion_matrix_df, metrics_df   
```

```python
confusion_matrix, metrics = performance_metrics(test_set_hat)
confusion_matrix
```
